## Explorando o DataSet: empréstimos de livros

```python
#Utilizando concat para unificar os dataframes de empréstimo

df_emprestimos = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10,
                            df11, df12, df13, df14, df15, df16, df17, df18,
                            df19, df20, df21], ignore_index=True)
```

```python
#A ordem das colunas está um pouco distante do fluxo "real".
#Coloquei em uma nova ordem e agora vamos começar uma breve etapa de limpeza.

nova_ordem_colunas = ['id_emprestimo', 'codigo_barras', 'matricula_ou_siape', 'tipo_vinculo_usuario', 'data_emprestimo',
                      'data_renovacao', 'data_devolucao']

df_emprestimos = df_emprestimos[nova_ordem_colunas]
```

```python
df_emprestimos.head()
```

```python
df_emprestimos.info()
```

```python
# Os campos de ID e Matrícula foram interpretados como números. Vou transformar em string.
# Os campos de data estão com o formato errado e vamos trasnformar para datetime.

df_emprestimos['matricula_ou_siape'] = df_emprestimos['matricula_ou_siape'].astype(str).str.rstrip('.0')
df_emprestimos['id_emprestimo'] = df_emprestimos['id_emprestimo'].astype(str)
df_emprestimos[['data_emprestimo', 'data_renovacao', 'data_devolucao']] = df_emprestimos[['data_emprestimo', 'data_renovacao', 'data_devolucao']].apply(pd.to_datetime)
```

```python
# Verificando valores duplicados

df_emprestimos.value_counts()
```

```python
# Removendo valores duplicados

df_emprestimos = df_emprestimos.drop_duplicates()
```

## Explorando o DataSet: acervo de livros

```python
df_exemplares.info()
```

```python
df_exemplares[['id_exemplar', 'registro_sistema']] = df_exemplares[['id_exemplar', 'registro_sistema']].astype(str)
```

```python
df_exemplares.head()
```

```python
df_exemplares.loc[(df_exemplares['localizacao'] >= 0) & (df_exemplares['localizacao'] <= 99), 'assunto_livro'] = 'Generalidades_ciencia_e_conhecimento'
df_exemplares.loc[(df_exemplares['localizacao'] >= 100) & (df_exemplares['localizacao'] <= 199), 'assunto_livro'] = 'Filosofia_psicologia'
df_exemplares.loc[(df_exemplares['localizacao'] >= 200) & (df_exemplares['localizacao'] <= 299), 'assunto_livro'] = 'Religiao'
df_exemplares.loc[(df_exemplares['localizacao'] >= 300) & (df_exemplares['localizacao'] <= 399), 'assunto_livro'] = 'Ciencias_sociais'
df_exemplares.loc[(df_exemplares['localizacao'] >= 400) & (df_exemplares['localizacao'] <= 499), 'assunto_livro'] = 'Classe_vaga'
df_exemplares.loc[(df_exemplares['localizacao'] >= 500) & (df_exemplares['localizacao'] <= 599), 'assunto_livro'] = 'Matematica_ciencias_naturais'
df_exemplares.loc[(df_exemplares['localizacao'] >= 600) & (df_exemplares['localizacao'] <= 699), 'assunto_livro'] = 'Ciencias_aplicadas'
df_exemplares.loc[(df_exemplares['localizacao'] >= 700) & (df_exemplares['localizacao'] <= 799), 'assunto_livro'] = 'Belas_artes'
df_exemplares.loc[(df_exemplares['localizacao'] >= 800) & (df_exemplares['localizacao'] <= 899), 'assunto_livro'] = 'Linguagem_lingua_linguistica'
df_exemplares.loc[(df_exemplares['localizacao'] >= 900) & (df_exemplares['localizacao'] <= 999), 'assunto_livro'] = 'Geografia_biografia_historia'
df_exemplares.loc[(df_exemplares['localizacao'].isna()), 'assunto_livro'] = 'Nao_catalogado'
```

```python
df_livros = df_exemplares.drop(['localizacao', 'registro_sistema'], axis=1)
df_livros.head()
```

## Merge entre os datasets

```python
df_merge = pd.merge(df_emprestimos, df_livros, on='codigo_barras', how='inner')

df_merge.head()
```

```python
df_merge.info()
```

```python
# Quantidade total de exemplares emprestados por cada ano

resposta01 = df_merge.groupby(df_merge['data_emprestimo'].dt.year)['id_exemplar'].count()
resposta01.plot(xlabel='Ano', ylabel='Livros', title='Total de Exemplares Emprestados | por Ano', marker='o')
plt.grid(True, which='both', linestyle='--', color='gray', linewidth=0.5)
plt.show()

```

```python
# Tempo médio de empréstimo dos livrospor categoria

tempo_medio = df_merge.groupby('assunto_livro').apply(lambda x:
 (x['data_devolucao'] - x['data_emprestimo']).mean().days).sort_values(ascending=False)

tempo_medio.plot(kind='bar', xlabel='Tema', ylabel='Dias de Leitura', title='Dias para devolução dos livros | por assunto')
plt.xticks(rotation=45, ha='right')
plt.show()
```
