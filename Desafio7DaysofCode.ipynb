## Quantidade total de exemplares emprestados por ano?

Quando falamos de exemplares emprestados precisamos calcular utilizando uma contagem de 'id_exemplar', já que um 'id_emprestimo' pode conter mais de um livro.

Notamos que há uma diminuição expressiva no empréstimo de livros nos primeiros meses de 2020, já que neste ano começávamos a experimentar o lockdown em função da Pandemia de Covid-19 (que teve início dem 11 de Março de 2020).

A esta pergunta de negócio acrescentamos uma média de tempo entre emprétimo e devolução para que a gestão da biblioteca avalie o tempo (médio) que um livro fica fora do acervo.

```python
# Quantidade total de exemplares emprestados por cada ano

resposta01 = df_merge.groupby(df_merge['data_emprestimo'].dt.year)['id_exemplar'].count()
resposta01.plot(xlabel='Ano', ylabel='Livros', title='Total de Exemplares Emprestados | por Ano', marker='o')
plt.grid(True, which='both', linestyle='--', color='gray', linewidth=0.5)
plt.show()

```

```python
# Tempo médio de empréstimo dos livrospor categoria

tempo_medio = df_merge.groupby('assunto_livro').apply(lambda x:
 (x['data_devolucao'] - x['data_emprestimo']).mean().days).sort_values(ascending=False)

tempo_medio.plot(kind='bar', xlabel='Tema', ylabel='Dias de Leitura', title='Dias para devolução dos livros | por assunto')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Quais os meses com menor demanda para programação de férias?

Uma mudança importante no direcionamento dos nossos cálculos: como estamos falando de demanda de trabalho, optei por considerar 'id_emprestimo', ou seja, a quantidade de "clientes" da biblioteca.

Para responder a esta pergunta de negócios comecei utilizando apenas 2019. Minha expectativa era entender se o último ano refletia o comportamento histórico dos usuários.

Chego a conclusão de que os melhores meses para férias são janeiro, dezembro, julho e junho. Meses que estão em linha com as férias escolares e entre períodos das nossas universidades.

```python
contagem = df_merge.groupby(df_merge['data_emprestimo'].dt.to_period('M'))['id_emprestimo'].count()
contagem_por_mes_2019 = contagem[contagem.index.year == 2019]
contagem_por_mes_2019
```

```python
contagem_por_mes_2019.plot(kind='bar', xlabel='Mes-Ano', ylabel='Volumetria', title='Volume de trabalho por mês')
plt.xticks(rotation=40, ha='right')
plt.show()
```

```python
contagem_mensal = df_merge.groupby(df_merge['data_emprestimo'].dt.month)['id_emprestimo'].count()
contagem_mensal
```

```python
contagem_mensal.plot(kind='bar', xlabel='Mes', ylabel='Volumetria', title='Volume de trabalho por mês')
plt.xticks(rotation=0, ha='right')
plt.show()
```
