## Quantidade total de exemplares emprestados por ano?

Quando falamos de exemplares emprestados precisamos calcular utilizando uma contagem de 'id_exemplar', já que um 'id_emprestimo' pode conter mais de um livro.

Notamos que há uma diminuição expressiva no empréstimo de livros nos primeiros meses de 2020, já que neste ano começávamos a experimentar o lockdown em função da Pandemia de Covid-19 (que teve início dem 11 de Março de 2020).

A esta pergunta de negócio acrescentamos uma média de tempo entre emprétimo e devolução para que a gestão da biblioteca avalie o tempo (médio) que um livro fica fora do acervo.

```python
# Quantidade total de exemplares emprestados por cada ano

resposta01 = df_merge.groupby(df_merge['data_emprestimo'].dt.year)['id_exemplar'].count()
resposta01.plot(xlabel='Ano', ylabel='Livros', title='Total de Exemplares Emprestados | por Ano', marker='o')
plt.grid(True, which='both', linestyle='--', color='gray', linewidth=0.5)
plt.show()

resposta_01_tabela = pd.DataFrame({'Livros': resposta01.index, 'Volumetria': resposta01.values})
resposta_01_tabela
```

```python
# Tempo médio de empréstimo dos livrospor categoria

tempo_medio = df_merge.groupby('assunto_livro').apply(lambda x:
 (x['data_devolucao'] - x['data_emprestimo']).mean().days).sort_values(ascending=False)

tempo_medio.plot(kind='bar', xlabel='Tema', ylabel='Dias de Leitura', title='Dias para devolução dos livros | por assunto')
plt.xticks(rotation=45, ha='right')
plt.show()

tempo_medio_tabela = pd.DataFrame({'Tema': tempo_medio.index, 'Dias de leitura': tempo_medio.values}).sort_values(by='Dias de leitura', ascending=False)
tempo_medio_tabela
```

## Explorando as variáveis categóricas

A proposta para explorarmos as variáveis categóricas implica em discutirmos volumetrias para tipos de vínculos dos usuários da biblioteca, coleções, bibliotecas e CDU.

Algumas dessas variáveis não foram apresentadas em gráficos visto a volumetria de um item. "Acervo circulante" tem uma representatividade muito grande e a visualização gráfica não seria o melhor caminho em minha opinião. Sendo assim, construí tabelas que podem ser exploradas pela audiência de forma interativa (clicando em "Suggest charts" no Collab).

No caso específico de emrpéstimos por tipo de vínculo do usuário usei o "percentual do total" para mostrar outras possibilidades de análise, além da contagem simples.

```python
# “Como se distribuem os empréstimos de exemplares pelos tipos de vínculo dos usuários?”

emprestimo_por_vinculo = df_merge.groupby(df_merge['tipo_vinculo_usuario'])['id_exemplar'].count().sort_values(ascending=False)

emprestimo_por_vinculo_table = pd.DataFrame({'Vinculo do usuario': emprestimo_por_vinculo.index, 'Total de empréstimos': emprestimo_por_vinculo.values})
emprestimo_por_vinculo
```

```python
percentual_total_vinculo = round(df_merge['tipo_vinculo_usuario'].value_counts(normalize=True) * 100,2)
percentual_total_vinculo.plot(kind='bar', xlabel='Vinculo', ylabel='Volumetria', title='% de Emprestimos por tipo de vinculo')
plt.xticks(rotation=45, ha='right')
plt.show()

percentual_total_vinculo = pd.DataFrame({'Coleção': percentual_total_vinculo.index, '% do Total': percentual_total_vinculo.values})

display(percentual_total_vinculo)
```

```python
# Quais coleções são mais emprestadas?

emprestimos_colecoes = df_merge.groupby(df_merge['colecao'])['id_exemplar'].count().sort_values(ascending=False)

emprestimos_colecoes_tabela = pd.DataFrame({'Coleção': emprestimos_colecoes.index, 'Volumetria': emprestimos_colecoes.values})
emprestimos_colecoes_tabela
```
